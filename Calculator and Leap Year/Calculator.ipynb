```python
logo = """
 _____________________
|  _________________  |
| | Pythonista   0. | |  .----------------.  .----------------.  .----------------.  .----------------. 
| |_________________| | | .--------------. || .--------------. || .--------------. || .--------------. |
|  ___ ___ ___   ___  | | |     ______   | || |      __      | || |   _____      | || |     ______   | |
| | 7 | 8 | 9 | | + | | | |   .' ___  |  | || |     /  \     | || |  |_   _|     | || |   .' ___  |  | |
| |___|___|___| |___| | | |  / .'   \_|  | || |    / /\ \    | || |    | |       | || |  / .'   \_|  | |
| | 4 | 5 | 6 | | - | | | |  | |         | || |   / ____ \   | || |    | |   _   | || |  | |         | |
| |___|___|___| |___| | | |  \ `.___.'\  | || | _/ /    \ \_ | || |   _| |__/ |  | || |  \ `.___.'\  | |
| | 1 | 2 | 3 | | x | | | |   `._____.'  | || ||____|  |____|| || |  |________|  | || |   `._____.'  | |
| |___|___|___| |___| | | |              | || |              | || |              | || |              | |
| | . | 0 | = | | / | | | '--------------' || '--------------' || '--------------' || '--------------' |
| |___|___|___| |___| |  '----------------'  '----------------'  '----------------'  '----------------' 
|_____________________|
"""
print(logo)

def add(n1, n2):
    return n1 + n2

def substract(n1, n2):
    return n1 - n2

def multiply(n1, n2):
    return n1 * n2

def divide(n1, n2):
    return n1 / n2

operations = {
    "+" : add,
    "-" : substract,
    "*" : multiply,
    "/" : divide,
}

num1 = float(input("Give first number: "))
num2 = float(input("Give second number: "))
new_number = True
while new_number:
    for symbol in operations:
        print(symbol)
    operation_symbol = input("Pick an operation from the above line: ")

    if operation_symbol == "+":
        answer = add(num1, num2)
    elif operation_symbol == "-":
        answer = substract(num1, num2)
    elif operation_symbol == "*":
        answer = multiply(num1, num2)
    elif operation_symbol == "/":
        answer = divide(num1, num2)
    else:
        print("Wrong Symbol")
    
    print(f"{num1} {operation_symbol} {num2} = {answer}")
    continue_op = input(f"Do you want to continue with {answer} ? Yes/No \n").lower()
    if continue_op == "yes":
        num1 = answer
        num2 = float(input("Give new number: "))
    elif continue_op == "no":
        num1 = float(input("Give first number: "))
        num2 = float(input("Give second number: "))
    else:
        new_number = False
```
